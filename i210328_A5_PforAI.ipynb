### Article Generation using N-grams

You have to write the urdu article using ngrams (1 grams to 5 grams). So, in short your output must be of 5 paragraphs, the first one is generated using unigram, second one is generated using bigram and so on.

Input: Your input is the seed sentence. E.g. first 3 to 4 words of the paragraph.

Output: Your output is the consist of 5 paragraphs for each n gram, each of 200 words.

You have to make N-gram model using the provided dataset. Dataset can be downloaded from  https://www.kaggle.com/datasets/saurabhshahane/urdu-news-dataset

You have to use all News Text column.

```python
#Your code here
import pandas as pd

newsText = pd.read_excel(r"C:\Users\hoxto\Desktop\urdu.xlsx", usecols = "C")
```

```python
# Split each paragraph into a list of words from the 'News Text' column
newsTextList = [x.split() for x in newsText['News Text'] if isinstance(x, str)]
```

```python
def ngramDictionary(n = 1):
    '''
    Returns an ngram dictionary based on the number of n given by user. Where the key is the ngram
    and the value is the ngram which appeared after it
    '''

    # For every sentence(list) in the newsTextList, create a list of strings which contain ngrams of 'n' of that sentence
    ngramList = [[" ".join(sentence[i:i+n]) for i in range(len(sentence) - n)] for sentence in newsTextList]

    # Now create tuple pairs of the current ngram and the ngram infront of it
    ngramTupleList = [(ls[i], ls[i+1]) for ls in ngramList for i in range(len(ls) - 1)]

    # Now to convert those tuple pairs into dictionary, set first element of tuple as key with an empty list as value
    # and then append the second element of the tuple into that list. Return the dictionary
    ngramDict = {}
    for key, value in ngramTupleList:
        ngramDict.setdefault(key, []).append(value)

    return ngramDict
```

```python
from random import choice

def sentence(wordCount, ngramCount):
    '''
    Returns a paragraph based on the word count and uses the ngramCount to create an ngram dictionary based on it
    '''

    # Basic checks to avoid a weird error
    if ngramCount < 1: return print("ngram must be a positive integer")
    if wordCount < ngramCount: return print("Word count must be bigger than ngram")

    # Paragraph is our final output, ngram is dictionary, currentGram is the ngram string picked from a random key
    # picked form ngram(Dictionary) variable. Append the key to the paragraph list
    ngram = ngramDictionary(ngramCount)
    paragraph = []
    currentGram = choice(list(ngram.keys()))
    paragraph.append(currentGram)

    # Loop until word count is reached. Pick a random value from currentGram(key)'s values and set it as the new key
    # Append the last word of the new key into the paragraph
    # If there is a key error due to lack of data in database, just pick another random key from the ngram dictionary
    while len(paragraph) < wordCount:
        try:
            currentGram = choice(ngram[currentGram])
        except:
            currentGram = choice(list(ngram.keys()))
        paragraph.append(currentGram.split()[-1])

    # Return the paragraph as an actual sentence by joining the lists with space
    return " ".join(paragraph)
```

```python
for i in range(5):
    print(f"Printing paragraph based on ngram = {i+1}:\n\n {sentence(200, i+1)}\n\n")
```

### Classify language out of the list given below using just stop words. Remove punctuations, make it lower.

```python
import nltk
from nltk.corpus import stopwords

def countLanguage(text):
    '''
    Count how many stopwords from different languages are present in text and
    outputs a dictionary showing the counts
    '''

    # Set of lowercase tokenized words. Does not include punctuation
    words =  { x.lower() for x in nltk.word_tokenize(text) if x.isalpha() }

    # Dictionary where keys are the languages and values are the length of the common
    # sets in words and set of stopwords.words(lang) using bitwise and operator
    d = { lang: len(set(stopwords.words(lang)) & words) for lang in stopwords.fileids() }

    return d

test = "An article is qualunque member van un class of dedicated words naquele estão used with noun phrases per mark the identifiability of the referents of the noun phrases"
countLanguage(test)
```

### Rule Based Roman Urdu Text Normalization

Roman Urdu lacks standard lexicon and usually many spelling variations exist for a given word, e.g., the word zindagi (life) is also written as zindagee, zindagy, zaindagee and zndagi. So, in this question you have to Normalize Roman Urdu words using the following Rules given in the attached Pdf. Your Code works for a complete Sentence or multiple sentences.

For Example: zaroori, zaruri, zarori map to the 'zrory'. So zrory becomes the correct word for all representations mentioned above.

```python
#Your code here

import nltk
import regex

# Tokenize the input sentence(s) and convert it to lowercased list
sentence = nltk.word_tokenize(input("Please enter your sentence: "))
sentence = [x.lower() for x in sentence]

# Apply the roman urdu lexican rules on each word. Over here, rule number precedence is also accounted for
for index, value in enumerate(sentence):
    # Rule number 1 ('ain' at the end replaced by 'ein')
    if 'ain' in value[-3:]:
        value = sentence[index] = value[:-3] + 'ein'

    # Rule number 2 ('ar' except at start replaced by 'r')
    if 'ar' in value:
        value = sentence[index] = value[0] + value[1:].replace('ar', 'r')

    #Rule number 3 ('ai' replaced by 'ae')
    if 'ai' in value:
        value = sentence[index] = value.replace('ai', 'ae')

    # Rule number 4 ('iy(y)' replaced by i)
    if 'iy' in value:
        while ('iy' in value):
            sentence[index] = value = value.replace('iy', 'i')

    # Rule number 5 ('ay' at the end replaced by 'e')
    if 'ay' in value[-2:]:
        value = sentence[index] = value[:-2] + 'e'

    # Rule number 6 ('ih(h)' replaced by 'eh')
    if 'ih' in value:
        while ('ihh' in value):
            value = value.replace('ihh', 'ih')
        value = sentence[index] = value.replace('ih', 'eh')

    # Rule number 7 ('ey' at the end replaced by 'e')
    if 'ey' in value[-2:]:
        value = sentence[index] = value[:-2] + 'e'

    # Rule number 8 ('s(s)' replaced by 's')
    if 'ss' in value:
        while ('ss' in value):
            sentence[index] = value = value.replace('ss', 's')

    # Rule number 9 ('ie' at the end replaced by 'y')
    if 'ie' in value[-2:]:
        value = sentence[index] = value[:-2] + 'y'

    # Rule number 10 ('ry' except at the end replaced by 'ri')
    if 'ry' in value:
        value = sentence[index] = value[:-1].replace('ry', 'ri') + value[-1:]

    # Rule number 11 ('es' at the start replaced by 'is')
    if 'es' in value[:2]:
        value = sentence[index] = 'is' + value[2:]

    # Rule number 12 ('sy' except at the end replaced by 'si')
    if 'sy' in value:
        value = sentence[index] = value[:-1].replace('sy', 'si') + value[-1:]

    # Rule number 13 ('a(a)' replaced by 'a')
    if 'aa' in value:
        while ('aa' in value):
            sentence[index] = value = value.replace('aa', 'a')

    # Rule number 14 ('ty' except at the end replaced by 'ti')
    if 'ty' in value:
        value = sentence[index] = value[:-1].replace('ty', 'ti') + value[-1:]

    # Rule number 15 ('j(j)' replaced by j)
    if 'jj' in value:
        while('jj' in value):
            sentence[index] = value = value.replace('jj', 'j')

    # Rule number 16 ('o(o)' replaced by o)
    if 'oo' in value:
        while('oo' in value):
            sentence[index] = value = value.replace('oo', 'o')

    # Rule number 17 ('ee(ee)' replaced by i)
    if 'ee' in value:
        while('ee' in value):
            sentence[index] = value = value.replace('ee', 'i')

    # Rule number 18 ('i' to 'y' if preceded by 'bcdefghijklmnopqrtuvwxyz')
    if 'i' in value[-1] and value[-2] in 'bcdefghijklmnopqrtuvwxyz':
        value = sentence[index] = value[:-1] + 'y'

    # Rule number 19 ('d(d)' replaced by 'd')
    if 'dd' in value:
        while('dd' in value):
            sentence[index] = value = value.replace('dd', 'd')

    # Rule number 20 ('u' replaced by 'o')
    if 'u' in value:
        value = sentence[index] = value.replace('u', 'o')

# Join the list with a space. If there is a space between a word and a punctuation, remove it and then print the sentence
sentence = regex.sub(r'\s([?.,!"])', r'\1', " ".join(sentence))
print(f"The sentence after normalization is: {sentence}")
```
